```python
import numpy as np

#缺失值处理

#缺失值
"""
nan:不是一个数字 1/0
None:python原生
pd.NA:数字不存在

"""
s=pd.Series([1,2,np.nan,None,pd.NA])
s
#查看缺失值
#print(s.isna())
#print(s.isnull())

df=pd.DataFrame([[1,2,np.nan],[3,4,5],[None,2,3]],columns=['第一列','第二列','第三列'])
#df.isnull().sum()


#缺失值处理
#1.剔除缺失值
s.dropna()
df.dropna(how='all')#剔除一整条记录 how='':指定提出的规则，全部为缺失值时删除
df.dropna(subset=['第一列'])#指定哪一列出现缺失值时删除




#填充缺失值
df=pd.read_csv('data/weather_withna.csv')


df.isna().sum()

#使用字典填充
df.fillna({'temp_max':20})

#使用统计值填充
df=df.fillna(df[['temp_max','temp_min','wind']].mean()).tail()

df.ffill().tail()#使用前面相邻的值填充

df.bfill().tail()#使用后面的值填充

df

```
