
### 查看完全重复的行（所有列都一模一样）
df.duplicated()          # 返回 True/False 系列
df.duplicated().sum()    # 看看一共有多少行完全重复

### 指定某些列看是否重复（最常用！）
df.duplicated(subset=['姓名'])           # 只看姓名是否重复
df.duplicated(subset=['身份证号'])       # 查重神器
df.duplicated(subset=['订单号', '日期'])

### 删除完全重复的行（保留第一次出现的）
df.drop_duplicates(inplace=True)

###只根据某些列去重（强烈推荐这样用）
df.drop_duplicates(subset=['姓名'], keep='first', inplace=True)   # 保留第一次出现
df.drop_duplicates(subset=['身份证号'], keep='last', inplace=True) # 保留最后一次
### 'first'  → 保留第一次出现（默认）
### 'last'   → 保留最后一次出现
### False    → 全部删除，连第一次也删（极少用）

```python
#数据分箱
"""
数据分箱:pd.cut(x,bins,labels)

x:数据
bins: bins=n 平均分成n段区间,起始值和结束值分别是最大值和最小值
"""

df=pd.read_csv(DATA/"employees.csv")
df1=df.head(10)[['employee_id','salary']]
df1
```

```python
pd.cut(df1['salary'],bins=3).value_counts()

df1['收入水平']=pd.cut(df1['salary'],bins=[0,10000,20000,30000],
labels=['低','中','高'])

df1

#平均分，每一段人数尽可能相同
pd.qcut(df1['salary'],q=3).value_counts()
```

分组聚合

```python
"""
分组聚合
df.groupby('分组依据')['聚合的字段'].聚合函数
"""

df=pd.read_csv(DATA/"employees.csv")
df.tail()
df=df.dropna(subset=['department_id'])
df['department_id']=df['department_id'].astype('int32')

df
#不同部门的薪资平均值
df2=df.groupby('department_id')[['salary']].mean()

df2['salary']=df2['salary'].round(2)
df2=df2.reset_index()
df2.sort_values('salary',ascending=False,inplace=True)
df2
```
