## 2.Series

创建Series

```python
from operator import index

#创建Series
import pandas as pd
from pandas import Series

#自定义索引
s=pd.Series([1,2,3,4,5],index=[1,2,3,4,5])
#定义标签
s=pd.Series([1,2,3,4,5],index=[1,2,3,4,5],name='月份')
print(s)

#2.通过字典创建
d={"a":1,"b":2,"c":3}
s=pd.Series(d)
print(s)

s1=pd.Series(s,index=["a","c"])
print(s1)
```

Series的属性

```python
#Series的属性
print(s.index)
print(s.values)
print(s.shape,s.ndim,s.size )

#显式索引
print(s.loc["a"])
#隐式索引
print(s.iloc[0])

print(s[:])
```

常用方法


```python
s2 = pd.Series([85, 92, 78, 95, 88],
               index=['张三', '李四', '王五', '赵六', '周七'],
               name='Python成绩')

#print(s2.name)

#找出成绩大于90的学生
#print(s2[s2>90])

#找出名字包含赵的

print(s2[s2.index.str.contains("赵")])
```

```python
math = pd.Series({'张三':92, '李四':88, '王五':75, '赵六':96})
python = pd.Series({'张三':85, '李四':92, '王五':78, '赵六':95, '周七':88})


total=math+python
print(total)
```

```python
# 学生成绩统计
import numpy as np

np.random.seed(42)

score=pd.Series(np.random.randint(50,100,10),index=["学生"+str(i) for i in range(1,11,1)])
print(score)
#计算平均分
avg=pd.Series.mean(score)
max=pd.Series.max(score)
min=pd.Series.min(score)
print(f"最高分{max},最低分{min},平均分{avg}")

#找出高于平均分的学生人数
print("高于平均分学生人数%d"%pd.Series.count(score[score>avg]))
```

## DataFrame

```python
data={'产品名称':['A','B','C','D'],
      '单价':[100,150,200,120],
      '销量':[50,30,20,40]
      }
data=pd.DataFrame(data)

data['总销售额']=data['单价']*data['销量']

data.nlargest(1,columns='总销售额')

print(data.sort_values(by='总销售额', ascending=False))
```

```python
data = {
    '用户ID': [101, 102, 103, 104, 105],
    '用户名': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    '商品类别': ['电子产品', '服饰', '电子产品', '家居', '服饰'],
    '商品单价': [1200, 300, 800, 150, 200],
    '购买数量': [1, 3, 2, 5, 4]
}

df = pd.DataFrame(data)

df['总消费金额']=df['商品单价']*df['购买数量']

df.nlargest(1,columns=['总消费金额'])

df['总消费金额'].mean()


df[df['商品类别']=="电子产品"]['购买数量'].sum()
```
