```python
"""
企鹅体重分析
目标：1.男性企鹅以及女性企鹅体重对比
    2.三种企鹅的体重有什么显著差异
"""
from pathlib import Path
DATA=Path(r"C:\Users\21471\PycharmProjects\JupyterProject\data")
```

```python
import numpy as np
import pandas as pd


#导入数据
penguin=pd.read_csv(DATA/"penguins.csv")

penguin.head()
```

```python
#数据清洗
penguin.isna().sum()
penguin.dropna(inplace=True)
penguin['is_male']=penguin['sex'].map({'Male':True,'Female':False})

penguin.groupby('is_male')['body_mass_g'].mean().round(2)
```

```python
penguin.groupby('species')['body_mass_g'].describe().round(2)
```

```python
#数据分箱，把体重分为低中高三个阶段

penguin['mess_level']=pd.cut(penguin['body_mass_g'],bins=3,labels=['低','中','高'])
penguin['mess_level'].value_counts()
```
